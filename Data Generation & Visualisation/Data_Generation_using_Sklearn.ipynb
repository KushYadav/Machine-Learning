## Scikit-Learn
Some useful functions to generate datasets available in sklearn.datasets are:
* make_regression
* make_classification
* make_blobs
* make_moons

```python
from sklearn.datasets import make_regression
from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

### Make Regression

```python
X,y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=5)
print(X.shape,y.shape)
plt.scatter(X,y)
plt.show()
```

### Make Classification

```python
X,y=make_classification(n_samples=400,n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1,n_classes=4)
print(X.shape,y.shape)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

### Make Blobs

```python
X,y=make_blobs(n_samples=400,n_features=2,centers=5,random_state=10)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

### Make Moons

```python
X,y=make_moons(n_samples=100,random_state=5,noise=.1)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```
