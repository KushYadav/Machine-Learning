## Types of Problems in Supervised Learing 

### Regression

Output is a continous value of examples.
* Predicting weather.
* Stock prices.
* Air quality.
* Predicting marks and so on..

### Classification

Output is discrete class of label.
* Apple of mango.
* Cat or not.
* Movie rating 1,2,3,4 or 5 given text of the review.

## Data for Classification

```python
u1=np.array([2,4])
cov1=np.array([[1,.7],[0.7,1]])
X1=np.random.multivariate_normal(u1,cov1,300)
u2=np.array([-1,-3])
cov2=np.array([[1,0],[0,5]])
X2=np.random.multivariate_normal(u2,cov2,400)
```

```python
plt.scatter(X1[:,0],X1[:,1],label='Apple')
plt.scatter(X2[:,0],X2[:,1],label='Mango',marker='^')
plt.xlim((-8,8))
plt.xlim((-8,8))
plt.legend()
plt.show()
```

```python
y=np.zeros(700)
y[:300]=1
X=np.vstack((X1,X2))
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
for i in range(700):
    if y[i]==1 :
        plt.scatter(X[i,0],X[i,1],color='Purple')
    else:
        plt.scatter(X[i,0],X[i,1],color='Grey')
plt.show()
```

```python
# or
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```
