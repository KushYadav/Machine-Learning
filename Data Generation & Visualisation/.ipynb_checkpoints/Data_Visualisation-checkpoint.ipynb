```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line and Scatter Plot
* If there is only one axis provided then it goes to the y-axis whereas the indices goes to the x-axis.

```python
x=np.arange(5)
print(x)
```

```python
plt.plot(x)
plt.show()
```

```python
x1=x**2
x2=x**3
plt.plot(x1,color='green')
plt.plot(x2)
plt.show()
```

```python
x1=x**2
x2=x**3
plt.scatter(x,x1,color='green',label='Legend')
plt.scatter(x,x2,marker="^")
plt.xlabel('xLabel')
plt.ylabel('yLabel')
plt.title('Title')
plt.legend()
plt.show()
```

## Bar Charts

```python
numbers1=np.random.randint(0,10,5)
indices1=np.arange(5)
numbers2=np.random.randint(0,10,5)
indices2=np.arange(5)
```

```python
plt.bar(indices1,numbers1,color='Orange',width=.4,label="Label_1")
plt.bar(indices2+.5,numbers2,width=.4,label="Label_1")
plt.legend()
plt.show()
```

## Pie Charts

```python
labels=['Algebra','Calculus','Coding','Algorithms']
values=[90,80,54,40]
plt.pie(values,labels=labels,radius=1.5,explode=(0,0,1,0),shadow=True)
plt.show()
```

## Histogram
* Standard normal function => mean=0, Standard Deviation=1 

```python
u=5
sigma=1

vals=u+sigma*np.random.randn(200)
plt.hist(vals,20)
plt.show()
```

```python
x=vals
y=np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```

```python
vals=np.round(vals)
z=np.unique(vals,return_counts=True)
print(z)
```

## Multivariate Normal Distribution

```python
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])
dist=np.random.multivariate_normal(mean,cov,500) # No. of points
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
mean=np.array([0.0,0.0])
cov=np.array([[1,.8],[.8,1]])
dist=np.random.multivariate_normal(mean,cov,500)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
mean1=np.array([0.0,0.0])
cov1=np.array([[1,.8],[.8,1]])
mean2=np.array([5,5])
cov2=np.array([[1.5,0],[0,1.5]])

dist1=np.random.multivariate_normal(mean1,cov1,500) 
dist2=np.random.multivariate_normal(mean2,cov2,500) 
plt.scatter(dist1[:,0],dist1[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```

## Subplots

```python
x=np.linspace(-np.pi,np.pi,50)


plt.subplot(2,2,1)
y1=np.sin(x)
plt.plot(x,y1)

plt.subplot(2,2,2)
y2=np.cos(x)
plt.plot(x,y2)

plt.subplot(2,2,3)
y3=np.tan(x)
plt.plot(x,y3)

plt.subplot(2,2,4)
y4=np.exp(x)
plt.plot(x,y4)

plt.show()
```
