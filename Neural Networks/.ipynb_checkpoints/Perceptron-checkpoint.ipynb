# Perceptron

- A Perceptron is a single layer neural network
- It acts a Linear Classifier, Binary classification
- Loss Function : Binary Cross Entropy
  (-ve of Log Likelihood)
- Optimisation : Gradient Descent/ Stochastic Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs,make_moons
```

## Data Generation

```python
X,y = make_blobs(n_samples=500, n_features=2, centers=2, random_state=3)
print(X.shape,y.shape)
```

```python
plt.style.use('seaborn')
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

## Functions

```python
def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))
```

```python
def predict(X,weights):
    z=np.dot(X,weights)
    return sigmoid(z)
```

```python
def loss(X,y,weights):
    y_=predict(X,weights)
    return np.mean(-y*np.log(y_)-(1-y)*np.log(1-y_))
```

```python
def update(X,y,weights,learning_rate):
    y_=predict(X,weights)
    dw=np.dot(X.T,y_-y)
    
    m=X.shape[0]
    weights=weights-learning_rate*dw/(float(m))
    return weights
```

```python
def train(X,y,learning_rate=.9,maxEpochs=100):
    ones=np.ones((X.shape[0],1))
    X=np.hstack((ones,X))
    
    weights=np.zeros(X.shape[1])
    
    for i in range(maxEpochs):
        weights=update(X,y,weights,learning_rate)
        
        if i%10==0:
            l=loss(X,y,weights)
            print("Epoch %d Loss %.4f"%(i,l))
            
    return weights        
```

```python
def get_predictions(X_test,weights,labels=True):
    if X_test.shape[1]!=weights.shape[0]:
        ones=np.ones((X_test.shape[0],1))
        X_test=np.hstack((ones,X_test))
        
    probs=predict(X_test,weights)
    
    if not labels:
        return probs
    else:
        labels=np.zeros(probs.shape)
        labels[probs>=.5]=1
        
    return labels
```

```python
weights=train(X,y)
```

```python
x1=np.linspace(-8,4,10)
print(x1)
x2=-(weights[0]+weights[1]*x1)/weights[2]
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.Accent)
plt.plot(x1,x2,c='red')
plt.show()
```

```python
y_=get_predictions(X,weights,labels=True)
score=np.sum(y_==y)/y.shape[0]
print(score)
```

## Moon Dataset

```python
X,y=make_moons(n_samples=500,noise=.2,random_state=3)
print(X.shape,y.shape)
```

```python
plt.style.use('seaborn')
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
weights=train(X,y)
```

```python
x1=np.linspace(-2,3,10)
print(x1)
x2=-(weights[0]+weights[1]*x1)/weights[2]
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.Accent)
plt.plot(x1,x2,c='red')
plt.show()
```

```python
y_=get_predictions(X,weights,labels=True)
score=np.sum(y_==y)/y.shape[0]
print(score)
```
