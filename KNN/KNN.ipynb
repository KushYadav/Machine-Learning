# K-Nearest Neighbours
* Supervised learning algorithm.
* Used for both regression and classification.
* No traing is involved.
* Used as a baseline algorithm.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
X,y=make_blobs(n_samples=500,n_features=2,centers=5,center_box=[-5,5],random_state=11)
print(X.shape,y.shape)
print(np.unique(y,return_counts=True))
```

```python
q_point=np.array([0,0])
plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(q_point[0],q_point[1],c='Red')
plt.xlim([-7.5,7.5])
plt.ylim([-7.5,7.5])
plt.show()

# Purple -> Dark Blue -> Cyan -> Light Green -> Yellow (0,1,2,3,4)
```

```python
def distance(a,b):
    return np.sqrt(sum((a-b)**2))
```

```python
def KNN(X,y,q_point,k=5):
    dis=[]
    
    for i in range(X.shape[0]):
        dis.append([distance(X[i],q_point),y[i]])
    
    dis=sorted(dis)
    dis=np.array(dis[:k])
    
    unq=np.unique(dis[:,1],return_counts=True)

    return unq[0][np.argmax(unq[1])]
```

```python
KNN(X,y,q_point)
```

```python
def accuracy(X_train,X_test,y_train,y_test,k=5):
    y_pred=[]
    
    for i in X_test:
        y_pred.append(KNN(X_train,y_train,i,k))
        
    return np.sum(y_pred==y_test)/X_test.shape[0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8)
```

```python
print(accuracy(X_train,X_test,y_train,y_test))
```
