```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles,make_moons
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
X,y = make_moons(n_samples=500,noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
k_means=KMeans(n_clusters=2)
k_means=k_means.fit(X)
centers=k_means.cluster_centers_
y_pred=k_means.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.scatter(centers[:,0],centers[:,1],marker="*",color='red')
plt.show()
```

## DBSCAN
* Density based spatial clustering of application with noise.
* Based upon the idea, that a cluster is a high density area sourrounded by low density region.
* Start by exploring a small area, if density is "good enough" it considered as a part of cluster, and explores neighbours and so on, so it groups regions based upon density.

```python
dbs=DBSCAN(eps=0.16,min_samples=4)
dbs=dbs.fit(X)
```

```python
y_pred=dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.show()
```
