```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)
y=y.reshape((-1,1))
print(X.shape)
print(y.shape)
```

```python
X=(X-X.mean())/X.std()
plt.scatter(X,y)
plt.title("Normalized Data")
plt.show()
```

```python
ones=np.ones((X.shape[0],1))
X_=np.hstack((X,ones))
```

```python
print(X_.shape)
print(X_[:5][:])
```

```python
def pred(X,theta):
    return np.dot(X,theta)
```

```python
def closed_form(X,y):
    
    y=np.mat(y)
    a=np.dot(X.T,X)
    b=np.dot(X.T,y)
    
    theta=np.linalg.pinv(a)*b
    return theta
```

```python
theta=closed_form(X_,y)
print(theta)
```

```python
X=(X-X.mean())/X.std()
plt.scatter(X,y)
plt.plot(X,pred(X_,theta),color='orange',label='Predicted')
plt.legend()
plt.title("Normalized Data")
plt.show()
```
